```python
import cv2
from PIL import Image
import numpy as np
```

```python
image = Image.open("bricks-jpg/2357 brick corner 1x2x2/2357 brick corner 1x2x2 000L.jpg")
image
```

```python
color = Image.new("RGB", image.size, (255,0,0))
mask = Image.new("RGBA", image.size, (0,0,0,160))
```

```python
Image.composite(image, color, mask)
```

```python
img = cv2.imread("bricks-jpg/2357 brick corner 1x2x2/2357 brick corner 1x2x2 000L.jpg")
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower = np.array([0,0,10])
upper = np.array([0,0,70])
```

```python
im_mask = cv2.inRange(hsv, lower, upper)
result = cv2.bitwise_and(hsv, hsv, mask=im_mask)
```

```python
hue = result[:,:,0]
```

```python
hue += 60
```

```python
result[:,:,0] = hue
```

```python
final = cv2.bitwise_or(cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB), img)
```

```python
cv2.imwrite("test.jpg", final)
```
